```python
from heapq import heappush, heappop
def dijkstra(g, src):
    rows = len(g)
    cols = len(g[0])
    
    graph = [[None]*rows for x in range(cols)]
    shortest_distances = [float("inf")] * (rows)
    shortest_distances[0] = 0
    
    for i in g:
        start, finish, cost = i
        graph[start][finish] = cost
        
    pq = [(0, src)]
    
    print("given graph =", graph)
    
    print("initial pq ", pq)
    
    while pq:
        curr_distance, curr_vertex = heappop(pq)
        
        print("HEAP POP:")
        print("curr_distance = ", curr_distance)
        print("curr_vertex = ", curr_vertex)
        
        for index, neighbour_value in enumerate(graph[curr_vertex]):
            
            print("graph[curr_vertex] = ", graph[curr_vertex])
            
            if neighbour_value:
                print("neighbour_value = ", neighbour_value)
                
                shortest_distances[index] = min(shortest_distances[index], shortest_distances[curr_vertex] + neighbour_value)
                
                print("shortest_distances[index] ", shortest_distances[index])
                
                heappush(pq, (shortest_distances[index], index))
                
                print("HEAP PUSH")
                print(pq)
                print("current shortest_distances ", shortest_distances)    
                print()

     
    print("shortest_distances ", shortest_distances)    
    
dijkstra([[0,1,100],[1,2,100],[0,2,500]], 0)



```
