```python
from collections import deque
class FlattenedIterator():
    
    def __init__(self, iterators):
        self.queue = deque()
        self.iterators = self.transform(iterators, self.queue)
        
    
    def transform(self, iterators_list, queue):
        length = len(iterators_list)
        cnt = 0 # counter to keep track if all subiterators are empty in the current iteration
        
        while cnt < length:
            cnt = 0 
            for it in iterators_list:
                curr = next(it, None)
                if curr is not None:
                    queue.append(curr)
                else:
                    cnt += 1
    
    def hasNext(self):
        return self.queue
    
    def nextElement(self):
        return self.queue.popleft()

def test_flattened_iterator(iterators):
    res = []
    flattened_iterator = FlattenedIterator(iterators)
    while(flattened_iterator.hasNext()):
        res.append(flattened_iterator.nextElement())
    return res
```

```python
def test_flattened_iterator(iterators):
    res = []
    flattened_iterator = FlattenedIterator(iterators)
    while(flattened_iterator.hasNext()):
        res.append(flattened_iterator.nextElement())
    return res    
```

```python
x = iter([1,2,3])
y = iter([4,5])
z = iter([6,7,8])
iterators = [x, y, z]
assert test_flattened_iterator(iterators) == [1, 4, 6, 2, 5, 7, 3, 8], "ERROR!"

x = iter([1,2,3])
iterators = [x]
assert test_flattened_iterator(iterators) == [1, 2, 3], "ERROR!"

x = iter([])
y = iter([4])
z = iter([6,7,8,9,10,11,12,13])
iterators = [x, y, z]
assert test_flattened_iterator(iterators) == [4, 6, 7, 8, 9, 10, 11, 12, 13], "ERROR!"

x = iter([])
y = iter([])
z = iter([6,7,8])
iterators = [x, y, z]
assert test_flattened_iterator(iterators) == [6, 7, 8], "ERROR!"

x = iter([3,4])
y = iter([])
z = iter([6,7,8, 9, 10])
iterators = [x, y, z]
assert test_flattened_iterator(iterators) == [3, 6, 4, 7, 8, 9, 10], "ERROR!"

x = iter([])
y = iter([])
iterators = [x, y]
assert test_flattened_iterator(iterators) == [], "ERROR!"


z = iter([])
iterators = [z]
assert test_flattened_iterator(iterators) == [], "ERROR!"


z = iter([-1])
iterators = [z]
assert test_flattened_iterator(iterators) == [-1], "ERROR!"
```
